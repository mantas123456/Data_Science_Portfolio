```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Show all columns and configure plot style
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

```python

# ✅ Correct relative path from the notebooks/ folder
DATA_PATH = "../data/Formula_1/"

# Load main CSV files
races = pd.read_csv(DATA_PATH + "races.csv")
results = pd.read_csv(DATA_PATH + "results.csv")
drivers = pd.read_csv(DATA_PATH + "drivers.csv")
constructors = pd.read_csv(DATA_PATH + "constructors.csv")
circuits = pd.read_csv(DATA_PATH + "circuits.csv")

# Preview the first few rows
display(races.head())
display(results.head())
display(drivers.head())
display(constructors.head())
display(circuits.head())
```

```python
for name, df in zip(["races", "results", "drivers", "constructors", "circuits"], [races, results, drivers, constructors, circuits]):
    print(f"\n{name.upper()} - Missing values:")
    display(df.isnull().sum())
```

```python
# Convert date column if needed
races['date'] = pd.to_datetime(races['date'])

# Count races per year
races['year'] = races['date'].dt.year
races_per_year = races['year'].value_counts().sort_index()

# Plot
plt.figure(figsize=(12, 4))
sns.barplot(x=races_per_year.index, y=races_per_year.values, color='skyblue')
plt.title("Number of F1 Races per Year")
plt.xlabel("Year")
plt.ylabel("Race Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
driver_counts = results['driverId'].value_counts().head(10)
top_driver_ids = driver_counts.index

top_drivers = drivers[drivers['driverId'].isin(top_driver_ids)].copy()
top_drivers['races_count'] = top_drivers['driverId'].map(driver_counts)

top_drivers = top_drivers[['driverRef', 'races_count']].sort_values(by='races_count', ascending=False)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(data=top_drivers, x='driverRef', y='races_count', palette='mako')
plt.title("Top 10 Drivers by Number of Races")
plt.ylabel("Number of Race Results")
plt.xlabel("Driver")
plt.tight_layout()
plt.show()
driver_counts = results['driverId'].value_counts().head(10)
top_driver_ids = driver_counts.index

top_drivers = drivers[drivers['driverId'].isin(top_driver_ids)].copy()
top_drivers['races_count'] = top_drivers['driverId'].map(driver_counts)

top_drivers = top_drivers[['driverRef', 'races_count']].sort_values(by='races_count', ascending=False)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(data=top_drivers, x='driverRef', y='races_count', palette='mako')
plt.title("Top 10 Drivers by Number of Races")
plt.ylabel("Number of Race Results")
plt.xlabel("Driver")
plt.tight_layout()
plt.show()
```
